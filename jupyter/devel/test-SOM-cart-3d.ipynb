```python
import pink
import numpy as np

som_dim = 2
neuron_dim = 3
channel_dim = 2

np_som = np.random.rand(som_dim, som_dim, neuron_dim, neuron_dim, channel_dim).astype(np.float32)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5))
plt.imshow(np_som.reshape((som_dim * som_dim, neuron_dim * neuron_dim * channel_dim)))
plt.colorbar()
plt.grid(False)
```

```python
som = pink.SOM(np_som, neuron_layout="cartesian-3d")
np_som_2 = np.array(som, copy=True)

plt.figure(figsize=(5,5))
plt.imshow(np_som_2.reshape((som_dim * som_dim, neuron_dim * neuron_dim * channel_dim)))
plt.colorbar()
plt.grid(False)
```

```python
print(np.allclose(np_som, np_som_2, rtol=1e-05, atol=1e-08, equal_nan=False))
diff = np.subtract(np_som, np_som_2)

plt.figure(figsize=(5,5))
plt.imshow(diff.reshape((som_dim * som_dim, neuron_dim * neuron_dim * channel_dim)))
plt.colorbar()
plt.grid(False)
```
